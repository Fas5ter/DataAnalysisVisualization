## A summary of the work developed, highlighting the importance of  the seaborn library.

Seaborn, a Python library, plays a critical role in data visualization by providing a high-level interface for creating informative and aesthetically pleasing graphs. Its importance lies in its ability to handle a wide range of chart types, facilitating comprehensive data exploration and analysis. This helps me as an analyst and data scientist by having good tools to work very well.
One of Seaborn's key strengths is its ability to create scatterplots and lineplots, allowing users to visualize relationships between variables effectively. Additionally, it excels in visualizing distributions through various plot types such as histplot, kdeplot, ecdfplot, and rugplot, enabling insights into data spread and density.

Moreover, Seaborn supports categorical data visualization with stripplot, swarmplot, boxplot, violinplot, pointplot, and barplot functionalities. These plots are essential for understanding the distribution of categorical variables and identifying patterns or trends within the data.

By offering a diverse range of plot types, Seaborn empowers analysts and researchers to gain deeper insights into their data effortlessly. Its intuitive syntax and extensive customization options make it a valuable tool for both exploratory data analysis and communicating findings effectively. Overall, Seaborn stands out as a crucial library for Python users seeking to visualize and understand complex datasets comprehensively.

```python
# Importamos seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
# Graficamos
sns.relplot(x=x, y = x**2, kind='scatter')
```

## Load Titanic dataset

```python
## Cargar Dataset Titanic
# Titanic Dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Vamos a crear cuatro diagramas de barras
fig, axes = plt.subplots(2,2, figsize=(12,8))
# Para el campo edad
sns.histplot(data=titanic, x='age', ax=axes[0,0])
# Para el grupo demogrÃ¡fico
sns.histplot(data=titanic, x='who', ax=axes[0,1], color='gold')
# Para la clase en la que viajan los pasajeros
sns.histplot(data=titanic, x='class', ax=axes[1,0], color='skyblue')
# Para conoce los puntos de embarque de los pasajeros
sns.histplot(data=titanic, x='embark_town', ax=axes[1,1], color='teal')
# Ponemos un titulo a la grÃ¡fica
fig.suptitle('Passengers Demographics', size= 24)
# Cambiemos el color de las grÃ¡ficas
```

```python
# Disribucion de la edad de los pasajeros segun la clase en la que viajan
# Vamos a sacar una funcion de distribucion a nivel grafica
sns.kdeplot(data=titanic, x='age')
```

```python
# Descomponiendo la distribucion segun la clase en la que viajan
sns.kdeplot(data=titanic, x='age', hue='class')
```

```python
ax = sns.kdeplot(data=titanic, x='age', hue='class', shade=True, palette='magma')
# Agregamos un titulo a la grÃ¡fica
ax.set_title('Pasenger Age By Class', size=20)
```

```python
# DistribuciÃ³n Acumulada de la Edad por Clase
ax = sns.ecdfplot(data=titanic, x='age', hue='class', palette='magma')
ax.set_title('Cumulative Age Distribution By Class', size=18)
```

```python
# Determinar si los pasajeros viajaban solos o acompaÃ±ados
sns.countplot(data=titanic, x='alone', palette='magma')
```

```python
sns.displot(data=titanic, x='who', palette='magma', hue='alone', multiple='stack')
```

```python
# Cuanto pagaron los viajeros
# En el eje x fare=precio del ticket y 30 bloques para agrupar las observaciones
sns.histplot(data=titanic, x='fare', bins=30)
```

```python
# Coloreamos por clases y alineamos los bloques
# El parametro hue nos permite colorear por clase
sns.histplot(data=titanic, x='fare', bins=30, hue='class', palette='magma', multiple='stack')
```

```python
# Does paying more help you chance to survive?
# Creamos una grafica categorica de caja por mostrar los que sobrevivieron y los que no
sns.catplot(data=titanic, x='alive', y='fare', palette='magma', col='class', kind='box', aspect=0.5)
```

```python
# Vamos a modificar la escala mediante una funciÃ³n logaritmica para quitar lo comprimido a las grÃ¡ficas
face_grid = sns.catplot(data=titanic.query('fare >0'), x='alive',  y='fare', col='class', kind='box', aspect=0.5, palette='magma')
face_grid.set(yscale='log')
```
